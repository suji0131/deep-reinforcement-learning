### Part 1: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def epsilon_greedy_policy(Q, state, no_of_actions, epsilon):
    """Takes a action-value fn, state, and generates an action to take in the state."""
    # generate which policy to 
    # choose "greedy"(exploitation) or "non-greedy"(exploration)
    policy_probs = [1-epsilon, epsilon]  # ["greedy"(exploitation), "non-greedy"(exploration)]
    policy_method = np.random.choice(np.arange(2), p=policy_probs)

    if policy_method == 0:  # "greedy"(exploitation)
        action = np.argmax(Q[state])
    elif policy_method == 1:  # "non-greedy"(exploration)
        probs = [1/no_of_actions for _ in range(no_of_actions)]
        action = np.random.choice(np.arange(no_of_actions), p=probs)
    else:
        raise ValueError("Policy generated is out-of-bounds.")

    return action
````

````python
def sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    # initialize performance monitor
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()   
        
        ## TODO: complete the function
        epsilon = 1 / i_episode  # i_episode starts from 1
        # no of actions 
        no_of_actions = env.action_space.n
        
        state = env.reset()
        action = epsilon_greedy_policy(Q, state, no_of_actions, epsilon)
        
        while True:
            # next state: S_{t+1} and reward: R_{t+1}
            next_state, reward, done, info = env.step(action)

            if done:
                # Q[next_state][next_action] is zero during the terminal state.
                Q[state][action] = Q[state][action] + alpha*(reward + 0 - Q[state][action])
                break
            else:
                # next action: A_{t+1}
                next_action = epsilon_greedy_policy(Q, next_state, no_of_actions, epsilon)
                Q[state][action] = Q[state][action] + alpha*(reward + gamma*(Q[next_state][next_action]) - Q[state][action])
            
                # reset state and action variable to next state and next action
                state = next_state
                action = next_action
        
    return Q
````

````python
# obtain the estimated optimal policy and corresponding action-value function
Q_sarsa = sarsa(env, 5000, .01)

# print the estimated optimal policy
policy_sarsa = np.array([np.argmax(Q_sarsa[key]) if key in Q_sarsa else -1 for key in np.arange(48)]).reshape(4,12)
check_test.run_check('td_control_check', policy_sarsa)
print("\nEstimated Optimal Policy (UP = 0, RIGHT = 1, DOWN = 2, LEFT = 3, N/A = -1):")
print(policy_sarsa)

# plot the estimated optimal state-value function
V_sarsa = ([np.max(Q_sarsa[key]) if key in Q_sarsa else 0 for key in np.arange(48)])
plot_values(V_sarsa)
````

### Part 2: TD Control: Q-learning

In this section, you will write your own implementation of the Q-learning control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def q_learning(env, num_episodes, alpha, gamma=1.0):
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        epsilon = 1 / i_episode  # i_episode starts from 1
        # no of actions 
        no_of_actions = env.action_space.n
        
        state = env.reset()
        action = epsilon_greedy_policy(Q, state, no_of_actions, epsilon)
        
        while True:
            # next state: S_{t+1} and reward: R_{t+1}
            next_state, reward, done, info = env.step(action)

            if done:
                # Q[next_state][next_action] is zero during the terminal state.
                Q[state][action] = Q[state][action] + alpha*(reward + 0 - Q[state][action])
                break
            else:
                Q[state][action] = Q[state][action] + alpha*(reward + gamma*(np.max(Q[next_state])) - Q[state][action])
            
                # reset state and action variable to next state and next action
                state = next_state
                # Now generate next action, A_{t+1}
                next_action = epsilon_greedy_policy(Q, next_state, no_of_actions, epsilon)
                action = next_action
        
    return Q
````

````python
def epsilon_greedy_policy_action_probabilities(Q, state, no_of_actions, epsilon) -> list:
    """Calculates the probability of selecting an action a when in state s, 
    for all a belongs to A(s)."""
    action_probabilities = [0] *  no_of_actions
    
    # choose "greedy"(exploitation) or "non-greedy"(exploration)
    # ["greedy"(exploitation), "non-greedy"(exploration)] -> [1-epsilon, epsilon]
    
    # when "greedy"(exploitation)
    action_idx = np.argmax(Q[state])
    action_probabilities[action_idx] = 1-epsilon
    
    # when "non-greedy"(exploration)
    # equiprobable action selection
    tmp_probs = [1/no_of_actions for _ in range(no_of_actions)]
    tmp_probs = [epsilon*i for i in tmp_probs]

    # Final action-probabilities
    action_probabilities = np.add(action_probabilities, tmp_probs)

    # print(np.sum(action_probabilities))
    # np.sum(action_probabilities) ~ 1.0
    
    return action_probabilities
````
